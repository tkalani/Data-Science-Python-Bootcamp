```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
# PLOT

fig = plt.figure(figsize=(4, 2), dpi=100) # size in inches figsize=(width, height)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
```

```python
# SUBPLOTS

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
# SAVING PLOTS TO FILES

fig.savefig('my_picture.png', dpi=200)
```

```python
# PLOT - LEGEND

fig = plt.figure() # size in inches figsize=(width, height)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X squared')
ax.plot(x, x**3, label='X cubed')

ax.legend(loc=0) # Labels are needed
ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
```
