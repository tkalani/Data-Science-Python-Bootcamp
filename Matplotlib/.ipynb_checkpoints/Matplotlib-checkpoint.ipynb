```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
print(x)
print(y)
```

```python
# FUNCTIONAL METHOD

plt.plot(x, y)
plt.show()

plt.plot(x, y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('TITLE')
```

```python
# MULTI-PLOT - FUNCTIONAL

plt.subplot(1, 2, 1) # rows, cols, plot_number
plt.plot(x, y, 'r-')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b-')
```

```python
# OBJECT-ORIENTED METHOD

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('TITLE')
```

```python
# OBJECT-ORIENTED METHOD

fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Y wrt X')

axes2.plot(y, x)
axes2.set_title('X wrt Y')
```

```python
# OBJECT-ORIENTED METHOD - SUBPLOTS

fig, axes = plt.subplots(nrows=1, ncols=2)
# print(axes)  JUST A TUPLE OF AXES
plt.tight_layout()

for current_ax in axes:
    current_ax.plot(x, y)
    
axes[0].plot(x, y, 'r-')
axes[0].set_title('TITLE 1')

axes[1].plot(y, x)
axes[1].set_title('TITLE 2')
```
