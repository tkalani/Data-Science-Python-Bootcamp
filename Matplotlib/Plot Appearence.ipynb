```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
# COLOURS

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#238c00')
ax.plot(y, x, color='purple')
```

```python
# LINEWIDTHS

# alpha -> transparency
# linewidth, lw -> width of line
# linestyle, ls -> style of line ('--', '-.', ':', 'steps')

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(y, x, color='purple', linewidth=20, alpha=0.5, linestyle='--')
```

```python
# MARKERS

# marker -> marks individual data points ('o', '+')
# markersize -> size of marker
# markerfacecolor
# markeredgecolor
# markeredgewidth

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(y, x, color='purple', marker='o', markersize=10,
       markerfacecolor='red', markeredgewidth=3, markeredgecolor='yellow')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_xlim([0, 1]) # lower_Bound, upper bound
ax.set_ylim([0, 2]) # lower_Bound, upper bound
```
